```python
from starDust2 import classify

import pickle
import os 

__current_dir = os.path.curdir

goodsnlc = pickle.load(open(__current_dir+'/../../kfo_hstlc/goodsnlc','rb'))

wheeler = goodsnlc[57]
wheeler.meta
```

```python
# 1.97 = [1.743,2.151]

classify_wheeler = classify.classify(wheeler,zhost=1.97,zhosterr = 0.2,zminmax=[1.743,2.151])
```

```python
pickle.dump(classify_wheeler,open('classify_wheeler.pkl','wb'))
```

```python
original_wheeler = classify.classify(wheeler,zhost=1.335,zhosterr=0.005,zminmax=[1.33,1.34])
```

```python
pickle.dump(original_wheeler,open('original_wheeler.pkl','wb'))
```

```python
original_wheeler = pickle.load(open('original_wheeler.pkl','rb'))

updated_wheeler = pickle.load(open('classify_wheeler.pkl','rb'))
```

```python
print(original_wheeler['bestmodel'])
print('original probabilities:pIa ~ {:.2e}, pIb/c ~ {:.2e}, pII ~ {:.2e}'.format(original_wheeler['pIa'],original_wheeler['pIbc'],original_wheeler['pII']))
print('log-likelihoods:{}'.format(original_wheeler['logztype']))
print(original_wheeler['snana-2006ez']['res']['param_dict'])
```

```python
# loglikelihoods not as good, agrees is still def a CC at updated further redshift 
print(updated_wheeler['bestmodel'])
print('original probabilities:pIa ~ {:.2e}, pIb/c ~ {:.2e}, pII ~ {:.2e}'.format(updated_wheeler['pIa'],updated_wheeler['pIbc'],updated_wheeler['pII']))
print('log-likelihoods:{}'.format(updated_wheeler['logztype']))
print(updated_wheeler['snana-2006ez']['res']['param_dict'])
```

```python
# original best fit as II

import sncosmo 

data = original_wheeler['snana-2006ez']['sn']
param_dict = original_wheeler['snana-2006ez']['res']['param_dict']
dust = sncosmo.CCM89Dust()
model = sncosmo.Model(source='snana-2006ez',
                      effects=[dust],effect_names=['host'],effect_frames=['rest'])
model.set(**param_dict)

sncosmo.plot_lc(data,model)
```

```python
# updated also best fit as II

import sncosmo 

data = updated_wheeler['snana-2006ez']['sn']
param_dict = updated_wheeler['snana-2006ez']['res']['param_dict']
dust = sncosmo.CCM89Dust()
model = sncosmo.Model(source='snana-2006ez',
                      effects=[dust],effect_names=['host'],effect_frames=['rest'])
model.set(**param_dict)

sncosmo.plot_lc(data,model)
```

```python
from starDust2 import classify

import pickle
import os 

__current_dir = os.path.curdir

goodsnlc = pickle.load(open(__current_dir+'/../../kfo_hstlc/goodsnlc','rb'))

biden = goodsnlc[0]
biden.meta
```

```python
# updated catalog specz = 1.525 (Skelton et al 2014) 
# https://iopscience.iop.org/article/10.1088/0067-0049/214/2/24/pdf

classify_biden = classify.classify(biden,zhost=1.525,zminmax=[1.53,1.52])
```

```python
pickle.dump(classify_biden,open('classify_biden.pkl','wb'))
```

```python
# original published specz = 0.585 +- 0.001 

original_biden = classify.classify(biden,zhost=0.585,zhosterr=0.001,zminmax=[0.580,0.590])
```

```python
pickle.dump(original_biden,open('original_biden.pkl','wb'))
```

```python
original_biden = pickle.load(open('original_biden.pkl','rb'))

updated_biden = pickle.load(open('classify_biden.pkl','rb'))
```

```python
print(original_biden['bestmodel'])
print('original probabilities:pIa ~ {:.2e}, pIb/c ~ {:.2e}, pII ~ {:.2e}'.format(original_biden['pIa'],original_biden['pIbc'],original_biden['pII']))
print('log-likelihoods:{}'.format(original_biden['logztype']))
print(original_biden['snana-2006ep']['res']['param_dict'])
```

```python
# improved likelihoods at the updated further redshift
# more confusion on type
print(updated_biden['bestmodel'])
print('update probabilities:pIa ~ {:.2e}, pIb/c ~ {:.2e}, pII ~ {:.2e}'.format(updated_biden['pIa'],updated_biden['pIbc'],updated_biden['pII']))
print('log-likelihoods:{}'.format(updated_biden['logztype']))
print(updated_biden['salt2-extended']['res']['param_dict'])
```

```python
# original best fit as Ib/c

import sncosmo 

data = original_biden['snana-2006ep']['sn']
param_dict = original_biden['snana-2006ep']['res']['param_dict']
dust = sncosmo.CCM89Dust()
model = sncosmo.Model(source='snana-2006ep',
                      effects=[dust],effect_names=['host'],effect_frames=['rest'])
model.set(**param_dict)

sncosmo.plot_lc(data,model)
```

```python
# update is best fit as Ia... though II not far behind

data = original_biden['salt2-extended']['sn']
param_dict = original_biden['salt2-extended']['res']['param_dict']

model = sncosmo.Model(source='salt2-extended')
model.set(**param_dict)

sncosmo.plot_lc(data,model)
```

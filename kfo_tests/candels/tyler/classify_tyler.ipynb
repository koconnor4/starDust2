```python
from starDust2 import classify

import os 
import pickle
__current_dir = os.path.curdir

egslc = pickle.load(open(__current_dir+'/../../kfo_hstlc/egslc','rb'))

tyler = egslc[12]
tyler.meta
```

```python
classify_tyler = classify.classify(tyler,zhost=0.683,zhosterr = 0.5,zminmax=[0.16,1.36])
```

```python
pickle.dump(classify_tyler,open('classify_tyler.pkl','wb'))
```

```python
original_tyler = classify.classify(tyler,zhost=1.95,zhosterr = 0.5,zminmax=[1.45,2.45])
```

```python
pickle.dump(original_tyler,open('original_tyler.pkl','wb'))
```

```python
original_tyler = pickle.load(open('original_tyler.pkl','rb'))

updated_tyler = pickle.load(open('classify_tyler.pkl','rb'))
```

```python
print(original_tyler['bestmodel'])
print('original probabilities:pIa ~ {:.2e}, pIb/c ~ {:.2e}, pII ~ {:.2e}'.format(original_tyler['pIa'],original_tyler['pIbc'],original_tyler['pII']))
print('log-likelihoods:{}'.format(original_tyler['logztype']))
print(original_tyler['snana-2006ep']['res']['param_dict'])
print(original_tyler['snana-2006ep']['res']['errors'])
```

```python
# improved likelihoods at updated nearer redshift
# type II more probable

print(updated_tyler['bestmodel'])
print('update probabilities:pIa ~ {:.2e}, pIb/c ~ {:.2e}, pII ~ {:.2e}'.format(updated_tyler['pIa'],updated_tyler['pIbc'],updated_tyler['pII']))
print('log-likelihoods:{}'.format(updated_tyler['logztype']))
print(updated_tyler['snana-2006ep']['res']['param_dict'])
print(updated_tyler['snana-2006ep']['res']['errors'])
```

```python
# original best fit as typeII

import sncosmo 

data = original_tyler['snana-2006ix']['sn']
param_dict = original_tyler['snana-2006ix']['res']['param_dict']
dust = sncosmo.CCM89Dust()
model = sncosmo.Model(source='snana-2006ix',
                      effects=[dust],effect_names=['host'],effect_frames=['rest'])
model.set(**param_dict)

sncosmo.plot_lc(data,model)
```

```python
# updated even better fit as typeII, but diff model

import sncosmo 

data = updated_tyler['snana-2006kv']['sn']
param_dict = updated_tyler['snana-2006kv']['res']['param_dict']
dust = sncosmo.CCM89Dust()
model = sncosmo.Model(source='snana-2006kv',
                      effects=[dust],effect_names=['host'],effect_frames=['rest'])
model.set(**param_dict)

sncosmo.plot_lc(data,model)
```

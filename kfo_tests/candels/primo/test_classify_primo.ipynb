```python
from starDust2 import classify

import pickle
import os 

__current_dir = os.path.curdir

goodsslc = pickle.load(open(__current_dir+'/../kfo_hstlc/goodsslc','rb'))

primo = goodsslc[16]

primo.meta
```

```python
classify_primo = classify.classify(primo,zhost=1.54992,zhosterr = 0.0001,zminmax=[1.549,1.551])
```

```python
# can plot marginal pdfs for any the best fit parameter values
# example time of peak...

import matplotlib.pyplot as plt


plt.plot(classify_primo['salt2-extended']['pdf']['t0'][0],classify_primo['salt2-extended']['pdf']['t0'][1])
plt.xlim(55460,55520)
plt.show()
print('mean ~ {}'.format(classify_primo['salt2-extended']['pdf']['t0'][2]), 'sigma ~ {}'.format(classify_primo['salt2-extended']['pdf']['t0'][3]))
```

```python
classify_primo['salt2-extended']['res'].keys()
```

```python
import sncosmo

param_dict = classify_primo['salt2-extended']['res']['param_dict']
model = sncosmo.Model(source='salt2-extended')
model.set(**param_dict)

data = classify_primo['salt2-extended']['sn']

sncosmo.plot_lc(data, model,zpsys='vega')
```

```python
# example bad result


param_dict = classify_primo['snana-2006ez']['res']['param_dict']
dust = sncosmo.CCM89Dust()

model = sncosmo.Model(source='snana-2006ez',
                      effects=[dust],effect_names=['host'],effect_frames=['rest'])
model.set(**param_dict)
model.set()
data = classify_primo['snana-2006ez']['sn']

sncosmo.plot_lc(data, model,zpsys='vega')
```
